```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
url = "https://api.github.com/repos/ironhack-datalabs/datamad1020-rev"
api_key = "token_ani"
```

```python
def requestGithub(url):
    token = os.getenv("token_ani")
    if not token:
        raise ValueError("Necesitas un Token")
        print(f"Haciendo request a {url}")
    
    params = {"Auhorization" : f"token {api_key}"}
    response = requests.get( url = url, params = params).json()
    
    return response
```

```python
respuesta = requestGithub(url)
print(respuesta)
```
